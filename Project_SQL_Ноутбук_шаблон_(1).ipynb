```python
# текст запроса
query5_1 = """
SELECT
rank,
employer_name,
vacancy_count
FROM (
    SELECT
        e.name as employer_name,
        COUNT(v.id) as vacancy_count,
        ROW_NUMBER() OVER (ORDER BY COUNT(v.id) DESC) as rank
    FROM vacancies v
    JOIN employers e ON v.employer_id = e.id
    GROUP BY e.id, e.name
) ranked_employers
WHERE rank IN (1, 5);
"""
```
